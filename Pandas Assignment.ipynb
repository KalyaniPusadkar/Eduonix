# USING PANDAS FUNCTION CALCULATE HOW MANY PRODUCTS COSTS MORE THAN $10?

```python
df.dtypes
```

```python
df['item_price'] = df['item_price'].str[1:].astype(float)
df['Per Item Price'] = df['item_price']/df['quantity']

```

```python
df
```

```python
item_above10= df.loc[(df['Per Item Price']>=10)]
```
